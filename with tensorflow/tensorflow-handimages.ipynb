```python
length = -1
for root, dirs, files in os.walk('train'):
    batch = []
    y = []
    length = len(files)
    for i in range(length):
        if i != 0 and i % 1000 == 0:
            batches.append(np.array(batch))
            batch = []
            y = np.array(y)
            Y.append(y)
            y = []
        image = Image.open('train/' + files[i])
        data = np.asarray(image)
        data = data.reshape((data.shape[0], data.shape[1], 1))
        batch.append(data)
        val = int(files[i][-6])
        answer = [val]
        '''
        for j in range(6):
            if j == val:
                answer.append(1)
            else:
                answer.append(0)
        '''
        y.append(answer)
    Y.append(y)
    if len(batch) != 0:
        batches.append(batch)
print("Done")
print("No. of training examples:", length)
```

```python
inputs = layers.Input((128, 128, 1))
x = layers.Conv2D(3, 4, padding = 'same', activation = 'relu')(inputs)
x = layers.MaxPooling2D(pool_size = (3, 3))(x)
x = layers.Conv2D(32, 3, padding = 'valid', activation = 'relu')(x)
x = layers.MaxPooling2D(pool_size = (3, 3))(x)
x = layers.Conv2D(64, 3, padding = 'valid', activation = 'relu')(x)
x = layers.MaxPooling2D(pool_size = (2,2))(x)
x = layers.Flatten()(x)
x = layers.Dense(128, activation = 'relu')(x)
x = layers.Dense(64, activation = 'relu')(x)
x = layers.Dense(32, activation = 'relu')(x)
outputs = layers.Dense(6)(x)

model = keras.Model(inputs = inputs, outputs = outputs)
```

```python
model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    optimizer = keras.optimizers.Adam(lr = 0.3),
    metrics = ['accuracy']
)
```

```python
X_test = [i for i in batches[-1]]
Y_test = []
X_train = []
Y_train = []
print(len(Y))

for i in range(len(batches[:-1])):
    for j in batches[i]:
        X_train.append(j)
    
    for j in Y[i]:
        Y_train.append(j)

for i in Y[-1]:
    Y_test.append(i)
print(len(X_train))
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
print(model.summary())
for i in model.layers:
    print(i.output_shape)
model.fit(X_train, Y_train, batch_size = 1000, epochs = 10, verbose = 2)
```

```python
model.evaluate(X_test, Y_test, batch_size = 1000, verbose = 2)
```
